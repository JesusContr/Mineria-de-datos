```python
import pandas as pd
import matplotlib.pyplot as plt

url= 'https://raw.githubusercontent.com/JesusContr/InteligenciaArtificial/main/FullData.csv'
df= pd.read_csv(url)

plt.figure(figsize=(100,100))
plt.subplot(1, 4, 1)
plt.scatter(df['Club'], df['Nationality'], c='red')
plt.title('JUGADOR', loc = "right", fontdict = {'fontsize':20, 'fontweight':'bold', 'color':'tab:red'})
plt.xlabel('Equipo', fontdict = {'fontsize':15, 'fontweight':'bold', 'color':'tab:red'})
plt.ylabel('Nacionalidad', fontdict = {'fontsize':15, 'fontweight':'bold', 'color':'tab:red'})

plt.figure(figsize=(200,200))
plt.subplot(1, 4, 2)
plt.scatter(df['Height'], df['Weight'], c='red')
plt.title('Medidas', loc = "right", fontdict = {'fontsize':20, 'fontweight':'bold', 'color':'tab:red'})
plt.xlabel('Altura', fontdict = {'fontsize':15, 'fontweight':'bold', 'color':'tab:red'})
plt.ylabel('Peso', fontdict = {'fontsize':15, 'fontweight':'bold', 'color':'tab:red'})

plt.figure(figsize=(200,200))
plt.subplot(1, 4, 3)
plt.scatter(df['Rating'], df['Age'], c='red')
plt.title('Nivel', loc = "right", fontdict = {'fontsize':20, 'fontweight':'bold', 'color':'tab:red'})
plt.xlabel('Media', fontdict = {'fontsize':15, 'fontweight':'bold', 'color':'tab:red'})
plt.ylabel('Edad', fontdict = {'fontsize':15, 'fontweight':'bold', 'color':'tab:red'})
```
