```python
import pandas as pd
import matplotlib.pyplot as plt
url= 'https://raw.githubusercontent.com/JesusContr/InteligenciaArtificial/main/FullData.csv'
df= pd.read_csv(url)

df.info()

sel_cols = ['Speed', 'Age','Dribbling','Agility']
Adulthood= ['Age', 'Speed']
a = df[sel_cols]
b = df[Adulthood]

def suma(df):
    sum = df.sum()
    return sum

def mean(df):
    mean = df.mean()
    return mean

def min(df):
    min = df.min()
    return min

def max(df):
    max = df.max()
    return max

def count(df):
    count = df.count()
    return count

def mode(df):
    mode = df.mode()
    return mode

def var(df):
    var = df.var()
    return var

def std(df):
    std = df.std()
    return std

def skew(df):
    skew = df.skew()
    return skew

df.info()
print('-------------------------------------------------------------------------------\n')

Age, Speed = b.apply(suma)
print ('La suma de las edades es:', Age,
       '\nLa suma de la velocidad es:', Speed)
print('-------------------------------------------------------------------------------\n')

Speed, Age, Dribbling, Agility = a.apply(max)
print('El maximo valor de edad es:', Age,
      '\nEl maximo valor de Dribles es:', Dribbling,
      '\nEl maximo valor de la Agilidad es:', Agility,
      '\nEl maximo valor de la velocidad es:', Speed)
print('-------------------------------------------------------------------------------\n')


Speed, Age, Dribbling, Agility = a.apply(min)
print('El valor minimo de edad es:', Age,
      '\nEl valor minimo de Dribles es:', Dribbling,
      '\nEl valor minimo de la Agilidad es:', Agility,
      '\nEl valor minimo de la velocidad es:', Speed)
print('-------------------------------------------------------------------------------\n')

Speed, Age, Dribbling, Agility = a.apply(count)
print('El conteo de edad es:', Age,
      '\nEl conteo de Dribles es:', Dribbling,
      '\nEl conteo de la Agilidad es:', Agility,
      '\nEl conteo de la velocidad es:', Speed)
print('-------------------------------------------------------------------------------\n')

Speed, Age, Dribbling, Agility = a.apply(mode)
print('La moda de edad es:', Age,
      '\nLa moda de los Dribles es:', Dribbling,
      '\nLa moda de la Agilidad es:', Agility,
      '\nLa moda de la velocidad es:', Speed)
print('-------------------------------------------------------------------------------\n')


Speed, Age, Dribbling, Agility = a.apply(mean)
print('La media de edad es:', Age,
      '\nLa media de los Dribles es:', Dribbling,
      '\nLa media de la Agilidad es:', Agility,
      '\nLa media de la velocidad es:', Speed)
print('-------------------------------------------------------------------------------\n')

Speed, Age, Dribbling, Agility = a.apply(var)
print('La varianza de edad es:', Age,
      '\nLa varianza de los Dribles es:', Dribbling,
      '\nLa varianza de la Agilidad es:', Agility,
      '\nLa varianza de la velocidad es:', Speed)
print('-------------------------------------------------------------------------------\n')


Speed, Age, Dribbling, Agility = a.apply(std)
print('La desviacion estandar de edad es:', Age,
      '\nLa desviacion estandar de los Dribles es:', Dribbling,
      '\nLa desviacion estandar de la Agilidad es:', Agility,
      '\nLa desviacion estandar de la velocidad es:', Speed)
print('-------------------------------------------------------------------------------\n')

Speed, Age, Dribbling, Agility = a.apply(skew)
print('La asimetria de edad es:', Age,
      '\nLa asimetria de los Dribles es:', Dribbling,
      '\nLa asimetria de la Agilidad es:', Agility,
      '\nLa asimetria de la velocidad es:', Speed)
print('-------------------------------------------------------------------------------\n')
```
