```python
import pandas as pd
import matplotlib.pyplot as plt
url= 'https://raw.githubusercontent.com/JesusContr/InteligenciaArtificial/main/FullData.csv'
df= pd.read_csv(url)
print(df.head(5))

df.info()

df.drop (columns=['Short_Pass','Long_Pass', 'Acceleration', 'Curve', 'Volleys', 'Freekick_Accuracy', 'Preffered_Foot', 'Work_Rate', 'Skill_Moves', 'Stamina', 'Sliding_Tackle', 'Standing_Tackle', 'Ball_Control', 'Marking', 'Reactions', 'Attacking_Position', 'Interceptions', 'Composure' ], inplace=True)
print('---------------------------------------------------------------------------------------------------\n',
      df.shape,
      '\n---------------------------------------------------------------------------------------------------')

df.info()

cols_cat = ['Nationality','Club','Club_Position', 'National_Position']
for col in cols_cat:
        print(f'Columna {col}: {df[col].nunique()} subniveles')

print('---------------------------------------------------------------------------------------------------\n',
      df.describe(),
      '\n---------------------------------------------------------------------------------------------------')

print(f'DF antes de eliminar las filas repetidas: {df.shape}')
df.drop_duplicates(inplace=True)
print(f'DF despu√©s de eliminar las filas repetidas: {df.shape}')

print('--------------------------------------------------------------------------------------------------------------------------------------------------------------\n',
      df.head(),
      '\n-----------------------------------------------------------------------------------------------------------------------------------------------------------')


```
